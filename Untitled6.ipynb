```python
grades2 = grades.copy()
grades2['Electronics 80'] = 100*(grades['Electronics 80']/80)
grades2['Signals 40'] = 100*(grades['Signals 40']/40)
grades2['Maths 40'] = 100*(grades['Maths 40']/40)
grades2['Microwaves 60'] = 100*(grades['Microwaves 60']/60)
grades2['Embedded 60'] = 100*(grades['Embedded 60']/60)
grades2['Enviroment(40)'] = 100*(grades['Enviroment(40)']/40)
grades2.head()
```

```python
grades2['Gender'].value_counts()
```

```python
ggg = grades2[['Gender','Electronics 80','Signals 40', 'Maths 40', 'Microwaves 60', 'Embedded 60', 'Enviroment(40)']].groupby('Gender')
ggg.head()
```

```python
ggg = pd.DataFrame(np.transpose(ggg.mean()))
ggg.index = ['Electronics','Signals','Maths','Microwaves','Embedded','Enviroment']
ggg
```

```python
ggg.plot(kind='bar',color=["magenta", "blue"])
plt.title("Mean scores based on Gender")
plt.ylabel("Mean Scores")
plt.legend(loc="upper center")
plt.show()
```

```python
male_mean = grades2[grades2['Gender'] == 'male']['SUM(320)'].mean()
feamale_mean = grades2[grades2['Gender'] == 'female']['SUM(320)'].mean()
print(f'Male mean: {male_mean}')
print(f'Female mean: {feamale_mean}')
print(f'Percentage Difference: {100*(feamale_mean - male_mean)/male_mean}')
```
